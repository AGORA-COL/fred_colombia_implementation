```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = '/data_HPC02/agora/fred_colombia_implementation/fred_input_files/imports/27_scaled_imports.csv'
```

```python
df_ = pd.read_csv(file)
```

```python
df_['Date'] = pd.to_datetime(df_['Date'])
```

```python
for var in ['Alpha', 'Gamma', 'Delta', 'Omicron']:
    df = df_[df_['variant'] == var]
    plt.plot(df.Date, df.scaled_imports, label=var)
plt.legend()
```
